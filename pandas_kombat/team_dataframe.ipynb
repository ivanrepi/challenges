### Kombat 3: How to get the items not common to both series A and series B?

Get all items of ser1 and ser2 not common to both.

```python
# Your code
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
```

```python
# Solution
a= ser1[ser1.isin(ser2) == False]
b= ser2[ser2.isin(ser1) == False]
notcommontoboth = pd.concat([a,b])
notcommontoboth
```

### Kombat 5: How to create a TimeSeries starting ‘2022-01-01’ and 10 weekends (saturdays) after that, having random numbers as values?

```python
# Your code

time_series = pd.Series(np.random.randint(1, 10, 10), )
```

```python
# Solution

array = pd.Series(pd.date_range("2022-01-01", freq="w", periods=10))
time_series = pd.Series(np.random.randint(1, 10, 10),index = array)

time_series
```
